1. Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

```python
import statistics
import matplotlib.pyplot as plt

# DATA TINGGI BADAN
data_tinggi_badan = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# MENGHITUNG MEAN
def count_mean(data):
    return sum(data) / len(data)

#  MENGHITUNG MODUS
def count_modus(data):
    modus = max(set(data), key = data.count)
    return modus

#  MENGHITUNG MEDIAN
def count_median(data):
    data.sort()
    n = len(data)
    if n % 2 == 0:
        median = (data[n//2 - 1] + data[n//2]) / 2
    else:
        median = data[n//2]
    return median

# MANUAL MENGHITUNG MEAN 
mean_manual = count_mean(data_tinggi_badan)

#  MANUAL MENGHITUNG MEDIAN
median_manual = count_median(data_tinggi_badan)

# MANUAL MENGHITUNG MEAN, MEDIAN, DAN MODUS 
modus_manual = count_modus(data_tinggi_badan)

# MENGHITUNG MEAN, MEDIAN, DAN MODUS DENGAN LIBRARY STATISTICS
median_stats = statistics.median(data_tinggi_badan)
mean_stats = statistics.mean(data_tinggi_badan)
modus_stats = statistics.mode(data_tinggi_badan)

# MENAMPILKAN HASIL
print(f"Mean (statistics): {mean_stats}")
print(f"Mean (Manual): {mean_manual}")
print(f"Modus (statistics): {modus_stats}")
print(f"Modus (Manual): {modus_manual}")
print(f"Median (statistics): {median_stats}")
print(f"Median (Manual): {median_manual}")

# VISUALISASI DISTRIBUSI DATA
plt.hist(data_tinggi_badan, bins=10, color='pink', edgecolor='black')
plt.axvline(mean_stats, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_stats, color='green', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(modus_stats, color='purple', linestyle='dashed', linewidth=2, label='Modus')
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```
